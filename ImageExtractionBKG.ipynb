```python
import os
import pandas as pd
import requests
from PIL import Image
from io import BytesIO

# Paths
csv_path = r"D:\downloads\ImageComp\bkg.csv"
save_base_dir = r"D:\downloads\ImageComp\PoC"

# Load the CSV file
df = pd.read_csv(csv_path)

# Ensure the required column exists
if 'hotel_id' not in df.columns:
    raise ValueError("The column 'hotel_id' is missing from the CSV file.")

# Filter columns: hotel_id and flw_urls[0] to flw_urls[162]
url_columns = [col for col in df.columns if col.startswith('flw_urls[')]
if not url_columns:
    raise ValueError("No URL columns (flw_urls[0] to flw_urls[162]) found in the CSV file.")

# Process each row in the DataFrame
for _, row in df.iterrows():
    hotel_id = row['hotel_id']
    if pd.isna(hotel_id):
        print(f"Skipping row with missing hotel_id.")
        continue

    # Create a directory for the hotel_id
    hotel_dir = os.path.join(save_base_dir, str(hotel_id))
    os.makedirs(hotel_dir, exist_ok=True)

    # Process URLs for the current row
    for url_col in url_columns:
        url = row[url_col]
        if pd.isna(url) or not isinstance(url, str):  # Skip if URL is missing or invalid
            continue

        try:
            # Download the image
            print(f"Processing URL: {url}")
            response = requests.get(url, stream=True, timeout=10)
            response.raise_for_status()  # Raise an error for bad status

            # Open the image
            img = Image.open(BytesIO(response.content))
            img_resized = img.resize((2400, 1600), Image.ANTIALIAS)

            # Infer content type or force default extension
            content_type = response.headers.get('Content-Type', '')
            extension = {
                'image/jpeg': '.jpg',
                'image/png': '.png',
                'image/gif': '.gif'
            }.get(content_type, '.jpg')  # Default to .jpg if type is unknown

            # Save the image
            original_name = f"{url_col}_{os.path.basename(url)}"  # Prefix with column name for uniqueness
            file_name = f"{os.path.splitext(original_name)[0]}{extension}"
            save_path = os.path.join(hotel_dir, file_name)
            img_resized.convert("RGB").save(save_path, "JPEG")

            print(f"Downloaded and saved: {save_path}")

        except Exception as e:
            print(f"Failed to process URL: {url}. Error: {e}")
```
